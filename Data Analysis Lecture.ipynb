## What are the countries in Europe?

Let's cleanup the data:

```python
countries_df['Region (Cleaned)'] = countries_df['Region'].map(str.strip)
```

```python
countries_df['Region'].unique()
```

```python
european = countries_df["Region"].isin(["WESTERN EUROPE", "EASTERN EUROPE"])
countries_df[european].filter(items=['Country', 'Region', 'Population'])
```

## Indexes

```python
countries_df.head()
```

```python
countries_df['Country'] = countries_df['Country'].map(str.strip)
countries_df = countries_df.set_index('Country')
```

```python
countries_df.at['France', 'Population']
```

```python
countries_df.filter(regex="^D", axis='rows')
```

```python
countries_df.sort_index(ascending=False)
```

```python
countries_df.sort_values(by='GDP ($ per capita)', ascending=False, na_position='first').head(5)
```

## Countries

```python
regions = countries_df.groupby('Region')
regions
```

```python
regions[['Population', 'Area (sq. mi.)']].sum().sort_values(by='Population', ascending=False)
```
