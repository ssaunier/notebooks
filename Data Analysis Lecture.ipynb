## European countries

Let's first clean our data (especially the column `Region`)

```python
countries_df["Region"] = countries_df["Region"].map(str.strip)
```

```python
european = countries_df["Region"].isin(["WESTERN EUROPE", "EASTERN EUROPE"])
```

```python
countries_df[european].head()
```

```python
countries_df[european].filter(items=["Country", "Region"])
```

## Re-indexing

```python
countries_df.head()
```

```python
countries_df = countries_df.set_index('Country')
```

```python
countries_df.head()
```

Let's select rows (countries) which start with the letter **D**:

```python
countries_df.filter(regex="^E", axis='rows')
```

## Sorting

```python
countries_df.sort_values('Population', ascending=False).head(10).filter(items=["Region", "Population"])
```

## Grouping rows

```python
regions = countries_df.groupby("Region")
regions
```

```python
top_regions = (regions[["Population"]].sum().sort_values("Population", ascending=False) / 1_000_000).round(1)
top_regions.columns = [ "Population (M)"]
top_regions
```

What **region** of the world is the **most dense**?

```python
regions[["Pop. Density (per sq. mi.)"]].mean().sort_values("Pop. Density (per sq. mi.)", ascending=False)
```

## Plotting a dataframe

```python
column = 'GDP ($ per capita)'
top_ten_countries_df = countries_df[[column]].sort_values(column, ascending=False).head(15)
```

```python
top_ten_countries_df
```
